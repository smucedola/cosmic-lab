```python
def x_hi(z, gamma):
    delta = 2.7*( gamma/1e-12)**2/3
    n_0 = 2e-7
    T = 1e4
    alpha = 4e-13
    
    return n_0*delta*(1+z)**3*alpha/gamma 
```

```python
gamma = np.array([0.0228, 0.0284, 0.0354, 0.0440,
                  0.0546, 0.0674, 0.0832, 0.102, 
                  0.125, 0.152, 0.185, 0.223, 0.267,
                  0.318, 0.376, 0.440, 0.510, 0.585,
                  0.660, 0.732, 0.799, 0.859, 0.909,
                  0.944, 0.963, 0.965, 0.950, 0.919,
                  0.875, 0.822])
gamma = gamma*1e-12

z = np.array([0, 0.05, 0.10, 0.16, 0.21, 0.27, 0.33, 0.40,
             0.47, 0.54, 0.62, 0.69, 0.78, 0.87, 0.96, 1.05,
             1.15, 1.26, 1.37, 1.49, 1.61, 1.74, 1.87, 2.01,
             2.16, 2.32, 2.48, 2.65, 2.83, 3.02])

plt.plot(z, x_hi(z, gamma));
plt.xlabel(r"$z$");
plt.ylabel(r"$X_{HI}$");
```
